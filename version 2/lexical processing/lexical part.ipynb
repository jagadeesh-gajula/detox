```python

import requests 
URL = "https://www.geeksforgeeks.org/data-structures/"
r = requests.get(URL) 



def clean(string):
    string=str(string)
    specials=['!','@','#','$','%','^','&','*','(',')','_','-','+','`','~',
              '?','|','/','"','=','<','>','{','}','[',']'
             ';',':','.',',','©','®','1','2','3','4','5','6','7','8','9','0']
    for special in specials:    
        string=string.replace(special," ")
    return string

#print(clean(r.content))
```

```python
from bs4 import BeautifulSoup

def spit(soup):
    sentence=[]
    soup = BeautifulSoup(r.content, 'html.parser')
    for i in soup.find_all('a'):
        i=str(i)
        beg=i.find('>')
        i=i[beg+1:]
        end=i.find('<')
        sentence.append(i[:end])
    return sentence
```
