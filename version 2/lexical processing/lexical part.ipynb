```python

import requests 
URL = "https://www.geeksforgeeks.org/data-structures/"
r = requests.get(URL) 



def clean(string):
    string=str(string)
    specials=['!','@','#','$','%','^','&','*','(',')','_','-','+','`','~',
              '?','|','/','"','=','<','>','{','}','[',']'
             ';',':','.',',','©','®','1','2','3','4','5','6','7','8','9','0']
    for special in specials:    
        string=string.replace(special," ")
    return string

#print(clean(r.content))
```

```python
from bs4 import BeautifulSoup

def spit(soup):
    sentence=[]
    soup = BeautifulSoup(r.content, 'html.parser')
    for i in soup.find_all('a'):
        i=str(i)
        beg=i.find('>')
        i=i[beg+1:]
        end=i.find('<')
        sentence.append(i[:end])
        
    for i in soup.find_all('title'):
        i=str(i)
        beg=i.find('>')
        i=i[beg+1:]
        end=i.find('<')
        sentence.append(i[:end])
        
    for i in soup.find_all('p'):
        i=str(i)
        beg=i.find('>')
        i=i[beg+1:]
        end=i.find('<')
        sentence.append(i[:end])
        
    return ' '.join(sentence)
```

```python
import pandas as pd
x=pd.read_csv('adult.csv')
```

```python
def protocol(url):
    if url.find('http') > -1:
        if url.find('https') > -1:
            return 'secure'
        else:
            return 'insecure'
    else:
        return 'protocol missing'
    
x['protocol']=x.sites.apply(lambda x:protocol(x))
```

```python
def extension(url):
    if protocol(url)!='protocol missing':
        url=url[10:]
        beg=url.find('.')
        end=url.find('/')
        return url[beg+1:end]
    else:
        return "not formatted"

x['extension']=x.sites.apply(lambda y:extension(y))
```

```python
def domain_name(url):
    if protocol(url)!= 'protocol missing':
        if url.find('www') == -1:
            beg=url.find('//')
            url=url[beg+2:]
            end=url.find('.')
            return url[:end]
        else:
            beg=url.find('www.')
            url=url[beg+4:]
            end=url.find('.')
            return url[:end]
    else:
        return 'protocol missing'
    
x['domain']=x.sites.apply(lambda x:domain_name(x))
```
